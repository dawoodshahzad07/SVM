# Task2

````python
from sklearn.model_selection import train_test_split 
````

````python
x=df.drop(['Class'],axis=1)
x.head()
````

````python
y=df['Class']
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
````

````python
from sklearn.svm import SVC
````

````python
model=SVC()
````

````python
model.fit(x_train,y_train)
````

````python
predict=model.predict(x_test)
predict
````

````python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
````

````python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,predict),columns=['Pred No','Pred Yes'],index=['Actual No','Actual yes'])
````

# Evaluating accuracy using different kernels

````python
#kernel=linear: usually used for large datasets
#C= regularization
model2=SVC(kernel='linear',C=0.1)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
````

````python
model2=SVC(kernel='linear',C=0.5)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
````

````python
#kernel=rbf: capturing complex, nonlinear relationships in the data
#gamma for choosing far or near instances
model2=SVC(kernel='rbf',gamma=0.1)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
````

````python
model2=SVC(kernel='poly')
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
````

````python
model2=SVC(kernel='sigmoid')
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
````
