```python
col=df.columns
for c in col:
    print(c,': ',df[c].nunique()) 
```

```python
#converting columns to numeric
df['Class']=df['Class'].str.replace('Kecimen','0')
df['Class']=df['Class'].str.replace('Besni','1')

df['Class']=df['Class'].apply(pd.to_numeric,errors='coerce')

df['Class'].dtype
```

```python
df['Class'].value_counts()
```

```python
for c in col: 
    print(c)
    plt.boxplot(df[c])
    plt.show()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x=df.drop(['Class'],axis=1)
x.head()
```

```python
y=df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model2=SVC(kernel='linear',C=0.1)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python
model2=SVC(kernel='linear',C=0.5)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python

model2=SVC(kernel='rbf',gamma=0.1)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python
model2=SVC(kernel='poly')
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python
model2=SVC(kernel='sigmoid')
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```
